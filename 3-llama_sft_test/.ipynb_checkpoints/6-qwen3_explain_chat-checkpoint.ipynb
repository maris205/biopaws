```python
import json
import random
import re
import os
import time
from datasets import load_dataset
from openai import OpenAI
from sklearn.metrics import accuracy_score, classification_report
```

```python
# ==========================================
# 1. 配置 API https://bailian.console.aliyun.com/#/home
# ==========================================
API_KEY = "sk-"  # 注意：在生产环境中请勿硬编码，建议使用 os.getenv
MODEL_ID = "qwen3-max"

client = OpenAI(
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
    api_key=API_KEY,
)
```

```python
# ==========================================
# 2. 数据准备 (保持不变)
# ==========================================
print("Loading dataset...")
try:
    # 加载数据集
    protein_data = load_dataset('dnagpt/biopaws', 'protein_pair_full')
    ds = protein_data['train']

    # 分离数据
    data_label_0 = [item for item in ds if item['label'] == 0]
    data_label_1 = [item for item in ds if item['label'] == 1]

    # 随机采样 (各50个)
    random.seed(42)
    sample_num = 50 
    sampled_0 = random.sample(data_label_0, sample_num)
    sampled_1 = random.sample(data_label_1, sample_num)

    # 合并并打乱
    combined_data = sampled_0 + sampled_1
    random.shuffle(combined_data)

    print(f"Data prepared: {len(combined_data)} pairs.")

except Exception as e:
    print(f"Error loading dataset: {e}")
    exit()

# 构建 Prompt 数据
prompt_data_list = []
id_to_ground_truth = {}

# 建立 ID 到具体数据的映射，方便后续第二轮追问时提取序列
id_to_data_content = {} 

for idx, item in enumerate(combined_data, 1):
    prompt_data_list.append({
        "id": idx,
        "seq_a": item['sentence1'],
        "seq_b": item['sentence2']
    })
    id_to_ground_truth[idx] = item['label']
    id_to_data_content[idx] = item # 存下原始数据
```

```python
# ==========================================
# 3. 第一轮：批量预测 (保持不变)
# ==========================================

# System Prompt
system_prompt = """You are an expert bioinformatics assistant capable of linguistic transfer learning.

1. The Concept:
In English, a sentence can be rearranged structurally but keep the same meaning (Paraphrase). Or, it can be scrambled to lose its logic (Adversarial).

2. The Analogy:
* **Homologous Proteins** are like **Paraphrases**: They have different sequences due to evolution, but they fold into the same structure and function.
* **Non-Homologous/Random Proteins** are like **Adversarial Sentences**: They look like proteins, but their internal structural logic is broken.

3. Output Requirements:
* I will provide a JSON list of protein pairs.
* You must return a RAW JSON object containing a list of results.
* The format must be strictly: `[{"id": 1, "prediction": "Homologous"}, {"id": 2, "prediction": "Non-Homologous"}, ...]`
* Do NOT provide explanations. Just the JSON array.
"""

# User Prompt 1
user_prompt_1 = f"""Here is the JSON list of 100 protein pairs to analyze.
Using your intuition about "sequence syntax" and "structural integrity," determine if each pair is "Homologous" or "Non-Homologous".

Data:
{json.dumps(prompt_data_list, indent=2)}
"""

# 初始化对话历史
messages_history = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt_1}
]

print("-" * 30)
print(f"ROUND 1: Calling Volcengine Model ({MODEL_ID}) for Batch Prediction...")

try:
    response_1 = client.chat.completions.create(
        model=MODEL_ID,
        messages=messages_history, # 使用历史列表
        temperature=0.1, 
        top_p=0.9,
    )

    full_content_1 = response_1.choices[0].message.content.strip()
    print("Round 1 Response received.")
    
    # 关键步骤：将模型的回复加入历史，形成上下文
    messages_history.append({"role": "assistant", "content": full_content_1})

except Exception as e:
    print(f"Round 1 API Call Failed: {e}")
    exit()
```

```python
# ==========================================
# 4. 第一轮评估 (保持不变)
# ==========================================
def parse_llm_json(text):
    try:
        match = re.search(r"\[.*\]", text, re.DOTALL)
        if match:
            return json.loads(match.group(0))
        return json.loads(text)
    except Exception:
        return [] 

predictions_list = parse_llm_json(full_content_1)
label_map = {"Homologous": 1, "Non-Homologous": 0}
y_true = []
y_pred = []

if predictions_list:
    for item in predictions_list:
        p_id = item.get('id')
        p_str = item.get('prediction')
        if p_id in id_to_ground_truth and p_str in label_map:
            y_true.append(id_to_ground_truth[p_id])
            y_pred.append(label_map[p_str])
    
    if y_true:
        acc = accuracy_score(y_true, y_pred)
        print(f"\nRound 1 Final Accuracy: {acc:.2%}")
else:
    print("Failed to parse Round 1 JSON.")
```

```python
# ==========================================
# 5. 第二轮：追问可解释性 (新增部分)
# ==========================================
print("\n" + "=" * 30)
print("ROUND 2: Interrogating the Model (Chain of Thought)...")
print("=" * 30)

# 我们挑选前 3 个样本让它解释（因为解释100个会导致 Token 溢出）
# 你也可以指定特定的 ID，比如挑一个错了的，或者挑 ID 1, 10, 20
example_ids = [1, 2, 3] 

user_prompt_2 = f"""Great job on the batch prediction. Your accuracy is impressive.

Now, I need to verify your "Chain of Thought". 
Please explain your reasoning for the first 3 pairs (ID: 1, 2, and 3) in detail.

For EACH pair, analyze:
1. **Motif Recognition**: Did you see any specific amino acid patterns (e.g., hydrophobic cores like L-V-I)?
2. **Structural Logic**: Why did you classify it as Homologous or Non-Homologous? 

Don't just say "it looks like it". Give me biological reasons.
"""

# 将新的追问加入历史
messages_history.append({"role": "user", "content": user_prompt_2})

try:
    response_2 = client.chat.completions.create(
        model=MODEL_ID,
        messages=messages_history, # 此时包含了 System -> User1 -> Assistant1 -> User2
        temperature=0.7, # 稍微提高温度，让它更愿意表达
        top_p=0.9,
    )

    full_content_2 = response_2.choices[0].message.content.strip()
    
    print("\n--- Model's Explanation (CoT) ---\n")
    print(full_content_2)
    
    # 验证一下：打印这几个 ID 的真实标签，看看它解释得对不对
    print("\n" + "-"*20 + " Ground Truth Verification " + "-"*20)
    for check_id in example_ids:
        print(f"ID {check_id}: True Label = {'Homologous' if id_to_ground_truth[check_id]==1 else 'Non-Homologous'}")
        print(f"       Seq A (First 20AA): {id_to_data_content[check_id]['sentence1'][:20]}...")

except Exception as e:
    print(f"Round 2 API Call Failed: {e}")

# 保存最终的对话记录，方便论文引用
with open("qwen3_full_dialogue.json", "w", encoding="utf-8") as f:
    json.dump(messages_history + [{"role": "assistant", "content": full_content_2}], f, indent=2, ensure_ascii=False)
    print("\nFull dialogue history saved to 'qwen3_full_dialogue.json'")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import textwrap

# ==========================================
# 论文专用：精修版 Case Study 可视化
# ==========================================

# 1. 准备数据
# 为了保证排版完美，这里我们"人工精选"了 CoT 中的核心语句
# 去掉了 Markdown 符号，精简了啰嗦的词汇，只保留 Kill Shot
cleaned_cot_text = """
[Motif Recognition]
• Seq A (Query):
  - Begins with signal peptide-like 'MRNQQL...'
  - Contains a Helix-Turn-Helix (HTH) segment.
  - Predicted function: Transcription Factor.

• Seq B (Target):
  - Features a metal-binding motif 'CVIRNY...'.
  - Resembles TIM-barrel enzymes (Enolase superfamily).
  - Lacks DNA-binding signatures found in Seq A.

[Structural Logic]
• The structural scaffolds are incompatible:
  - Seq A relies on HTH for DNA interaction.
  - Seq B forms a closed TIM-barrel barrel for catalysis.
• Despite low identity, the functional motifs diverge.

[Conclusion]
• Different evolutionary origins.
• Prediction: Non-Homologous.
"""

# 真实序列 (截取)
real_seq_a = "MRNQQLPIVPLLRDELPLDTVELARFMIGKYLVRDLPEGRMSG..." 
real_seq_b = "MLKQTLLYTGKAKSVYETDNADHLILVFRDDASAFNGEKIEQL..." 

# 2. 绘图设置
fig = plt.figure(figsize=(16, 7), dpi=150) # 宽一点，高一点
gs = fig.add_gridspec(1, 3, width_ratios=[1, 1.2, 0.8], wspace=0.05) # 调整列宽比例

ax_left = fig.add_subplot(gs[0])
ax_mid = fig.add_subplot(gs[1])
ax_right = fig.add_subplot(gs[2])

# ========================
# 左图：Input Sequences
# ========================
ax_left.axis('off')
ax_left.set_title("Input: Raw Sequences", fontsize=14, fontweight='bold', pad=20)

# 背景框
rect = patches.Rectangle((0.05, 0.3), 0.9, 0.5, linewidth=1, edgecolor='#bdc3c7', facecolor='#f4f6f7', zorder=-1)
ax_left.add_patch(rect)

# 文本
ax_left.text(0.1, 0.75, "Sequence A (Query):", fontsize=11, fontweight='bold', color='#2c3e50')
ax_left.text(0.1, 0.68, textwrap.fill(real_seq_a, width=35), fontsize=9, family='monospace', color='#555')

ax_left.text(0.1, 0.50, "Sequence B (Target):", fontsize=11, fontweight='bold', color='#2c3e50')
ax_left.text(0.1, 0.43, textwrap.fill(real_seq_b, width=35), fontsize=9, family='monospace', color='#555')


# ========================
# 中图：LLM Reasoning (CoT)
# ========================
ax_mid.axis('off')
ax_mid.set_title("Model Reasoning (Chain-of-Thought)", fontsize=14, fontweight='bold', pad=20)

# 绘制每一行，检测关键词进行高亮
y_pos = 0.85
line_height = 0.045

lines = cleaned_cot_text.strip().split('\n')
highlights = ["Helix-Turn-Helix", "TIM-barrel", "Non-Homologous", "Transcription Factor", "Enolase"]

for line in lines:
    if not line.strip(): 
        y_pos -= line_height/2 # 空行少空一点
        continue
    
    # 默认样式
    color = '#2c3e50'
    weight = 'normal'
    fontsize = 10
    
    # 标题样式 ([...])
    if line.startswith("["):
        weight = 'bold'
        fontsize = 11
        color = '#2980b9' # 蓝色标题
        y_pos -= 0.01
    
    # 关键词高亮
    is_highlight = False
    for kw in highlights:
        if kw in line:
            # 绘制整行 (简化处理，实际可以用 split 拼接实现局部变色)
            ax_mid.text(0.05, y_pos, line, fontsize=fontsize, family='monospace', color='#c0392b', fontweight='bold')
            # 荧光笔底色
            rect = patches.Rectangle((0.03, y_pos-0.01), 0.94, 0.035, linewidth=0, facecolor='#f9e79f', alpha=0.4, zorder=-1)
            ax_mid.add_patch(rect)
            is_highlight = True
            break
            
    if not is_highlight:
        ax_mid.text(0.05, y_pos, line, fontsize=fontsize, family='monospace', color=color, fontweight=weight)
    
    y_pos -= line_height

# ========================
# 右图：Structure Verification
# ========================
ax_right.axis('off')
ax_right.set_title("Structural Fact Check", fontsize=14, fontweight='bold', pad=20)

# 上方框：HTH
rect_a = patches.FancyBboxPatch((0.15, 0.65), 0.7, 0.15, boxstyle="round,pad=0.02", linewidth=2, edgecolor='#3498db', facecolor='white')
ax_right.add_patch(rect_a)
ax_right.text(0.5, 0.74, "Seq A Structure", ha='center', fontsize=10, fontweight='bold')
ax_right.text(0.5, 0.69, "Helix-Turn-Helix\n(DNA Binding)", ha='center', fontsize=9, color='#e74c3c', fontweight='bold')

# 下方框：TIM-barrel
rect_b = patches.FancyBboxPatch((0.15, 0.30), 0.7, 0.15, boxstyle="round,pad=0.02", linewidth=2, edgecolor='#f39c12', facecolor='white')
ax_right.add_patch(rect_b)
ax_right.text(0.5, 0.39, "Seq B Structure", ha='center', fontsize=10, fontweight='bold')
ax_right.text(0.5, 0.34, "TIM-barrel\n(Metabolic Enzyme)", ha='center', fontsize=9, color='#e74c3c', fontweight='bold')

# 中间的不等号
ax_right.text(0.5, 0.53, "≠", ha='center', va='center', fontsize=40, color='#95a5a6', alpha=0.5)
ax_right.text(0.5, 0.15, "Result: Non-Homologous", ha='center', fontsize=12, fontweight='bold', color='#c0392b', 
              bbox=dict(facecolor='#fadbd8', edgecolor='none', boxstyle='round,pad=0.5'))

plt.tight_layout()
plt.show()
```
