```python
# 只有当你希望生成一个独立的、巨大的模型文件时才跑这个
from peft import PeftModel
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
from huggingface_hub import login # 新增


# 1. 登录 Hugging Face (解决 403 权限问题)
HF_TOKEN = "" # 使用你之前提供的 Token
login(token=HF_TOKEN)



base_model_id = "meta-llama/Llama-3.1-8B"
adapter_path = "./Llama-3.1-8B-PAWS-En-Finetuned"
save_path = "./Llama-3.1-8B-PAWS-Merged-Full" # 合并后的保存路径

# 1. 加载模型 (注意：合并时不能用 4bit/8bit 加载，必须用 float16)
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    torch_dtype=torch.float16,
    device_map="auto"
)
model = PeftModel.from_pretrained(base_model, adapter_path)

# 2. 合并权重
print("Merging weights...")
model = model.merge_and_unload()

# 3. 保存完整模型
print(f"Saving full model to {save_path}...")
model.save_pretrained(save_path)
tokenizer = AutoTokenizer.from_pretrained(base_model_id)
tokenizer.save_pretrained(save_path)
```
