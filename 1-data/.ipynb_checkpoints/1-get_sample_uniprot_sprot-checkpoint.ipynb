## 获得uniprot_sprot采样蛋白质序列

```python
!wget ftp://ftp.uniprot.org/pub/databases/uniprot/current_release/knowledgebase/complete/uniprot_sprot.fasta.gz
```

```python
!gunzip uniprot_sprot.fasta.gz
```

```python
from Bio import SeqIO
import random

input_file = "uniprot_sprot.fasta"
output_file = "sampled_10000_proteins.fasta"

# 读取所有序列（内存足够的话）
all_sequences = list(SeqIO.parse(input_file, "fasta"))

# 随机抽样
sampled = random.sample(all_sequences, 10000)

# 保存
SeqIO.write(sampled, output_file, "fasta")
print(f"采样完成：{len(sampled)} 条序列")
```
