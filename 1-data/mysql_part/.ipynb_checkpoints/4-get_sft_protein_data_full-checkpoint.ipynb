```python
import pymysql
import pandas as pd

# ==================== æ•°æ®åº“é…ç½® ====================
HOST = ""
PORT = 3306
USER = ""
PASSWORD = ""
DATABASE = "ionet"
CHARSET = "utf8mb4"

# ==================== å‚æ•° ====================
POSITIVE_SAMPLES = 10000   # æ­£æ ·æœ¬ç›®æ ‡æ•°é‡
NEGATIVE_SAMPLES = 10000   # è´Ÿæ ·æœ¬ç›®æ ‡æ•°é‡
LENGTH_DIFF_MAX = 50       # ä¸€å¯¹åºåˆ—é•¿åº¦å·®ä¸è¶…è¿‡ 50 aaï¼ˆå…³é”®é™åˆ¶ï¼‰
OUTPUT_CSV = "protein_pair_20k_no_length_limit_but_diff_limited.csv"

# ==================== è¿æ¥æ•°æ®åº“ ====================
connection = pymysql.connect(
    host=HOST,
    port=PORT,
    user=USER,
    password=PASSWORD,
    database=DATABASE,
    charset=CHARSET
)

try:
    print(f"æ­£åœ¨æŠ½å–æ­£æ ·æœ¬ï¼ˆlabel=1ï¼‰ï¼Œç›®æ ‡ {POSITIVE_SAMPLES} æ¡ï¼Œä»…é™åˆ¶é•¿åº¦å·® â‰¤ {LENGTH_DIFF_MAX} aa...")
    pos_df = pd.read_sql("""
        SELECT sentence1, sentence2, label
        FROM protein_pair
        WHERE label = 1
          AND ABS(LENGTH(sentence1) - LENGTH(sentence2)) <= %s
        ORDER BY RAND()
        LIMIT %s
    """, connection, params=(LENGTH_DIFF_MAX, POSITIVE_SAMPLES))

    print(f"æ­£åœ¨æŠ½å–è´Ÿæ ·æœ¬ï¼ˆlabel=0ï¼‰ï¼Œç›®æ ‡ {NEGATIVE_SAMPLES} æ¡ï¼Œä»…é™åˆ¶é•¿åº¦å·® â‰¤ {LENGTH_DIFF_MAX} aa...")
    neg_df = pd.read_sql("""
        SELECT sentence1, sentence2, label
        FROM protein_pair
        WHERE label = 0
          AND ABS(LENGTH(sentence1) - LENGTH(sentence2)) <= %s
        ORDER BY RAND()
        LIMIT %s
    """, connection, params=(LENGTH_DIFF_MAX, NEGATIVE_SAMPLES))

    # æ£€æŸ¥å®é™…æŠ½å–æ•°é‡
    actual_pos = len(pos_df)
    actual_neg = len(neg_df)
    print(f"\nå®é™…æŠ½å–å®Œæˆï¼š")
    print(f"  æ­£æ ·æœ¬ (label=1): {actual_pos} æ¡")
    print(f"  è´Ÿæ ·æœ¬ (label=0): {actual_neg} æ¡")
    print(f"  æ€»æ ·æœ¬: {actual_pos + actual_neg} æ¡")

    if actual_pos < POSITIVE_SAMPLES:
        print(f"  è­¦å‘Šï¼šæ­£æ ·æœ¬ä¸è¶³ï¼Œä»…æŠ½åˆ° {actual_pos} æ¡ï¼ˆå¯èƒ½æ•°æ®åº“ä¸­é•¿åº¦å·® â‰¤ {LENGTH_DIFF_MAX} çš„æ ·æœ¬è¾ƒå°‘ï¼‰")
    if actual_neg < NEGATIVE_SAMPLES:
        print(f"  è­¦å‘Šï¼šè´Ÿæ ·æœ¬ä¸è¶³ï¼Œä»…æŠ½åˆ° {actual_neg} æ¡")

    # åˆå¹¶å¹¶æ‰“ä¹±
    sample_df = pd.concat([pos_df, neg_df], ignore_index=True)
    sample_df = sample_df.sample(frac=1, random_state=42).reset_index(drop=True)

    # ä¿å­˜ CSV
    sample_df.to_csv(OUTPUT_CSV, index=False, encoding="utf-8")

    print(f"\nğŸ‰ å®Œæˆï¼å…±ç”Ÿæˆ {len(sample_df)} æ¡æ ·æœ¬ï¼ˆä¸é™åˆ¶ç»å¯¹é•¿åº¦ï¼Œä»…é™åˆ¶ä¸€å¯¹é•¿åº¦å·® â‰¤ {LENGTH_DIFF_MAX} aaï¼‰")
    print(f"CSV æ–‡ä»¶å·²ä¿å­˜è‡³ï¼š{OUTPUT_CSV}")
    print("\nå‰5è¡Œé¢„è§ˆï¼š")
    print(sample_df.head())

    # å¯é€‰ï¼šè¾“å‡ºé•¿åº¦ç»Ÿè®¡ï¼Œä¾¿äºè®ºæ–‡æè¿°æ•°æ®é›†åˆ†å¸ƒ
    print("\né•¿åº¦ç»Ÿè®¡ï¼ˆä¾›è®ºæ–‡å‚è€ƒï¼‰ï¼š")
    print("sentence1 é•¿åº¦åˆ†å¸ƒï¼š")
    print(sample_df['sentence1'].str.len().describe())
    print("\nsentence2 é•¿åº¦åˆ†å¸ƒï¼š")
    print(sample_df['sentence2'].str.len().describe())
    print("\né•¿åº¦å·®åˆ†å¸ƒï¼š")
    print(abs(sample_df['sentence1'].str.len() - sample_df['sentence2'].str.len()).describe())

except Exception as e:
    print(f"é”™è¯¯ï¼š{e}")
finally:
    connection.close()
    print("æ•°æ®åº“è¿æ¥å·²å…³é—­ã€‚")
```
